```python
import matplotlib.pyplot as plt

hari = ['H1', 'H2', 'H3', 'H4', 'H5', 'H6', 'H7']
jumlah_penjualan = [40, 80, 110, 35, 20, 15, 25]


plt.figure(figsize=(8, 4))
plt.plot(hari, jumlah_penjualan, marker='*', color='g', linestyle='-')
plt.title('Line Chart Penjualan per hari')
plt.xlabel('hari')
plt.ylabel('jumlah penjualan')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

nama = ['ojak', 'bunga', 'cakra', 'nisa', 'aina', 'ismi', 'diven', 'ferdy', 'fita', 'rahma', 'arnold']
nilai = [90, 70, 85, 80, 95, 75, 100, 65, 78, 89, 60]

# Scatter Plot #1
plt.figure(figsize=(6, 8))
sns.scatterplot(x=nama, y=nilai)
plt.title('Hubungan antara nama dengan nilai')
plt.xlabel('nama')
plt.ylabel('nilai')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

luas_tanah = np.array([150, 200, 250, 300, 400])
harga_rumah = np.array([200, 350, 400, 550, 600])

X_train, X_test, y_train, y_test = train_test_split(luas_tanah.reshape(-1,1), harga_rumah, test_size=0.2, random_state=40)
model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
r_squared = model.score(X_test, y_test)
print("koefisien determinasi (R^2):", r_squared)

plt.figure(figsize=(6, 8))
plt.scatter(luas_tanah, harga_rumah, color='green', label='data')
plt.plot(luas_tanah, model.predict(luas_tanah.reshape(-1, 1)), color='blue', linewidth=3, label='Regresi Linear')
plt.xlabel('Luas tanah (m^2)')
plt.ylabel('Harga Rumah (juta Rupiah)')
plt.title('Prediksi Harga Rumah berdasarkan Luas Tanah')
plt.legend()
plt.grid(True)
plt.show
```
