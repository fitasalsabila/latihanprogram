```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
g = 9.80665  # m/s^2
L = 1  # m
Y = 0.25  # m
omega = 2.5  # rad/s

# Define the differential equation
def theta_dot(theta, t):
    return -g/L * np.sin(theta) + (omega**2)/L * Y * np.cos(theta) * np.sin(omega*t)

# Time points for plotting
t = np.linspace(0, 10, 1000)

# Initial condition: theta(0) = 0 (at rest)
theta_0 = 0

# Use Euler's method to solve the differential equation
theta = np.zeros_like(t)
theta[0] = theta_0
for i in range(1, len(t)):
    delta_t = t[i] - t[i-1]
    theta[i] = theta[i-1] + theta_dot(theta[i-1], t[i-1]) * delta_t

# Plot theta_dot versus t
plt.plot(t, theta_dot(theta, t))
plt.xlabel('Time (s)')
plt.ylabel('Theta Dot (rad/s)')
plt.title('Theta Dot versus Time')
plt.grid(True)
plt.show()

# Find the maximum value of theta_dot
max_theta_dot = np.max(theta_dot(theta, t))
print("Maximum theta_dot during t = 0 to 10 seconds:", max_theta_dot)
```
